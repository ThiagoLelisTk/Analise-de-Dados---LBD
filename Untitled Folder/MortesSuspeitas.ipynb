```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
path = 'Dados/'
startName = 'dados_de_'
filenames = {'abril2016', 'abril2017', 'abril2018', 'fevereiro2016', 'fevereiro2017', 'fevereiro2018', 'janeiro2016', 'janeiro2017', 'janeiro2018','marco2016', 'marco2017', 'marco2018'}
extension = '.csv'
```

```python
def populaDataFrame():
    dfList = list()
    
    for file in filenames:
        df = pd.DataFrame(pd.read_csv(path+startName+file+extension, sep =';'))
        dfList.append(df)
        
    df = pd.concat(dfList)
    return df
```

```python
df = populaDataFrame()
df.sort_values(by=['ANO_BO', 'DATAOCORRENCIA'])
```
